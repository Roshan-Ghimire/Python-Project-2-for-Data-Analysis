# Data Cleaning

```python
data.isnull().sum()
```

```python
data.columns=data.columns.str.strip()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
#Removing dollor sidn,and commo from columns so that we can convert it into float,int etc and perform aggregrate operation
dollor=['Units Sold','Manufacturing Price','Sale Price','Gross Sales','Discounts','Sales','COGS','Profit']
for columns in dollor:
    data[columns]=data[columns].replace('[\$,]','',regex=True)
```

```python
data.head()
```

```python
data.dtypes
```

```python
#replacing - with 0
data['Discounts']=data['Discounts'].replace('-','0',regex=True)
```

```python
data.head()
```

```python
# for converting the object data type to float we can use astype function
```

```python
changingtype=['Units Sold','Manufacturing Price','Sale Price','Gross Sales','Discounts','Sales','COGS']
for columns in changingtype:
    data[columns]=data[columns].astype(float)
```

```python
 data.dtypes
```

```python
# removing the extra white space present in profit column
data['Profit']=data['Profit'].str.strip()
```

```python
#replacing the round bracket with '' ie empty string and converting empty string into nan value
data['Profit']=data['Profit'].replace(r'[()]','', regex = True).replace('', pd.NA)
```

```python
data['Profit'] = pd.to_numeric(data['Profit'],errors='coerce')
```

```python
data['Profit'].unique()
```

```python
# change the datatype od date column 
data['Date']=pd.to_datetime(data['Date'])
```

```python
#droping column
data.drop(['Month Number','Month Name','Year'],axis=1,inplace=True)
```

```python
data.dtypes
```

# Data cleaning process is completed . Now we perform data analysis and visualization

# Country Wise Analysis

1) Calculate the Sales,Profit,Units sold,Cogs with every every country

```python
country_data=data.groupby('Country').agg({'Sales':sum,'Profit':sum,'Units Sold':sum,'COGS':sum}).reset_index()
country_data
```

```python
sb.set(rc={'figure.figsize':(18,5)})
Salesdata=sb.barplot(x='Country',y='Sales',data=country_data)
for bar in Salesdata.containers:
    Salesdata.bar_label(bar)
```

We can see that highest sale is done is USA and lowest sale is recorded in Mexico

```python
# calculate the unit sold to every country
sb.set(rc={'figure.figsize':(18,5)})
UnitSoldData=sb.barplot(x='Country',y='Units Sold',data=country_data)
for bar in UnitSoldData.containers:
    UnitSoldData.bar_label(bar)
```

We can see that highest Units  is sold to Canada and lowest Units Sold is recorded in German Country

```python
# Calculate the country wise profit
```

```python
#Since the Sales is more than COGS .so, we can say that company is in profit
Profitdata = data.groupby('Country')['Profit'].sum().reset_index()
Profitdata
```

```python
plt.figure(figsize=(10,6))
plt.pie(Profitdata['Profit'], labels=Profitdata['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Country wise Profit')
```
